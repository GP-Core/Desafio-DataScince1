#1. Análisis de facturación



```python
#librerias utlizadas
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
#Se procesa la facturación de las 4 tiendas
ventasPtienda1 = sum(tienda['Precio'])
ventasPtienda2 = sum(tienda2['Precio'])
ventasPtienda3 = sum(tienda3['Precio'])
ventasPtienda4 = sum(tienda4['Precio'])
```

```python
#Creando el grafico
#Parte horizontal del grafico
tiendas = ['Tienda 1','Tienda 2','Tienda 3', 'Tienda 4']
#parte vertical del grafico (ventas)
ventas = [ventasPtienda1,ventasPtienda2,ventasPtienda3,ventasPtienda4]

#Colores para el estilo
colores = ['#A6033F','#F285B8','#067302','#022601']
coloresBordes = ['#022601','#067302','#F285B8','#A6033F']

#determinamos el tamaño de la grafica
plt.plot(figsize=(9, 5))
#Graficacion
plt.bar(tiendas,ventas,color=colores,width=0.5,edgecolor=coloresBordes,linewidth=1.5)

#Etiquetas y titulos
plt.title('VENTAS TOTALES POR TIENDA')
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.grid(visible=True,which='major',axis='y',linestyle='--',linewidth=0.5,color='#444444') #para el lineado de la grafica

#Mostrar el numero de ventas
for i,ven in enumerate(ventas):#in enumerate nos permite conocer el indice y el valor que tiene
  plt.text(i,ven,f"${ven:,.0f}",ha='center',va='bottom')#primero va el x, luego el y, despues el valor y ya queda la decoracion


plt.savefig('ventas.png')
plt.show()
plt.close()

```

# 2. Ventas por categoría

```python
#librerias utlizadas
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Contamos productos por categoría para cada tienda
categorias_tienda1 = tienda['Categoría del Producto'].value_counts()
categorias_tienda2 = tienda2['Categoría del Producto'].value_counts()
categorias_tienda3 = tienda3['Categoría del Producto'].value_counts()
categorias_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Creamos un DataFrame combinando todas las tiendas
df_categorias = pd.DataFrame({
    'Tienda 1': categorias_tienda1,
    'Tienda 2': categorias_tienda2,
    'Tienda 3': categorias_tienda3,
    'Tienda 4': categorias_tienda4
}).fillna(0)

# Graficaamos
colores = ['#6E7346','#B4D9CE','#1F3140','#69731F']
vc = df_categorias.plot(kind='bar', figsize=(14, 7), width=0.8,color=colores)

# Título centrado
vc.set_title('Cantidad de productos vendidos por categoría en cada tienda', fontsize=16, pad=20)

# Etiquetas de los ejes
vc.set_xlabel('Categoría del Producto', fontsize=12)
vc.set_ylabel('Cantidad de Productos Vendidos', fontsize=12)

# Rotación de las categorías
plt.xticks(rotation=45, ha='right')
# Agregar los valores encima de cada barra sin que se empalmen
for bar in vc.patches:
    height = bar.get_height()
    if height > 0:
        vc.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5),  # separación vertical
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=8, color='black')

# Mostrar leyenda
vc.legend(title='Tiendas')



# Mostrar gráfica
plt.savefig('ventasCategorias.png')
plt.show()
```

# 4. Productos más y menos vendidos

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


def analizar_tienda_top5(df, nombre_tienda):
    conteos = df['Producto'].value_counts()

    # Top 5 más vendidos
    top_5_mas = conteos.head(5)

    # Top 5 menos vendidos
    top_5_menos = conteos.tail(5)

    resultados = []

    # Agregamos los 5 más vendidos
    for i, (producto, ventas) in enumerate(top_5_mas.items(), 1):
        resultados.append({
            'Tienda': nombre_tienda,
            'Tipo': 'Más Vendido',
            'Ranking': i,
            'Producto': producto,
            'Ventas': ventas
        })

    # Agregamos los 5 menos vendidos
    for i, (producto, ventas) in enumerate(top_5_menos.items(), 1):
        resultados.append({
            'Tienda': nombre_tienda,
            'Tipo': 'Menos Vendido',
            'Ranking': i,
            'Producto': producto,
            'Ventas': ventas
        })

    return resultados

# Analizamos todas las tiendas
tiendas_df = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda1', 'Tienda2', 'Tienda3', 'Tienda4']

todos_resultados = []
for df, nombre in zip(tiendas_df, nombres_tiendas):
    resultados_tienda = analizar_tienda_top5(df, nombre)
    todos_resultados.extend(resultados_tienda)

# Creamos el DataFrame final
df_resultado = pd.DataFrame(todos_resultados)

df_resultado
```

```python
#librerias utilizadas
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

#Sacamos los promedios de envio de cada tienda
promedioEnvioT1 = round(tienda['Costo de envío'].mean(),2)
promedioEnvioT2 = round(tienda2['Costo de envío'].mean(),2)
promedioEnvioT3 = round(tienda3['Costo de envío'].mean(),2)
promedioEnvioT4 = round(tienda4['Costo de envío'].mean(),2)

costos = {'Tienda1': promedioEnvioT1,'T2': promedioEnvioT2,'T3':promedioEnvioT3,'t4':promedioEnvioT4}
#Creacion del grafico
tiendas = ['Tienda 1','Tienda 2','Tienda 3', 'Tienda 4']
calificaciones = [promedioEnvioT1,promedioEnvioT2,promedioEnvioT3,promedioEnvioT4]
colores = ['#BF7E84','#D9D8D7','#734D3F','#8C3F3F']

#determinamos el tamaño de la grafica
plt.figure(figsize=(9, 5))

#Graficacion combinada
plt.plot(tiendas, calificaciones, color='#8C3F3F', linewidth=2, alpha=0.7)
plt.scatter(range(len(tiendas)), calificaciones, c=colores, s=150, alpha=0.9, edgecolors='black', zorder=5)

#Etiquetas y titulos
plt.title('COSTO PROMEDIO DE ENVIO POR TIENDA')
plt.xlabel('Tiendas')
plt.ylabel('Costo')

#Mostrar el numero de ventas
for i,calif in enumerate(calificaciones):
    plt.text(i,calif+0.5,f"{calif}",ha='center',va='bottom', fontweight='bold')

plt.grid(True, alpha=0.3)
plt.tight_layout()

plt.savefig('costoEnvio.png')
```
